```python
import numpy as np
import matplotlib.pyplot as plt

NF=np.loadtxt("./datas/orginal/fault_data_0HP_NF.txt")
plt.figure(figsize=(10, 3))
sig = NF[:8192]
plt.plot(sig, color='black')

IF=np.loadtxt("./datas/orginal/fault_data_0HP_IF.txt")
plt.figure(figsize=(10, 3))
sig = IF[:8192]
plt.plot(sig, color='black')

OF=np.loadtxt("./datas/orginal/fault_data_0HP_OF.txt")
plt.figure(figsize=(10, 3))
sig = OF[:8192]
plt.plot(sig, color='black')

BF=np.loadtxt("./datas/orginal/fault_data_0HP_BF.txt")
plt.figure(figsize=(10, 3))
sig = BF[:8192]
plt.plot(sig, color='black')

IOF=np.loadtxt("./datas/orginal/fault_data_0HP_IO.txt")
plt.figure(figsize=(10, 3))
sig = IOF[:8192]
plt.plot(sig, color='black')

IBF=np.loadtxt("./datas/orginal/fault_data_0HP_IB.txt")
plt.figure(figsize=(10, 3))
sig = IBF[:8192]
plt.plot(sig, color='black')

OBF=np.loadtxt("./datas/orginal/fault_data_0HP_OB.txt")
plt.figure(figsize=(10, 3))
sig = OBF[:8192]
plt.plot(sig, color='black')

IOBF=np.loadtxt("./datas/orginal/fault_data_0HP_IOB.txt")
plt.figure(figsize=(10, 3))
sig = IOBF[:8192]
plt.plot(sig, color='black')
```

```python
dim=2048
step=100
categorie_num=0

x_train=[]

for i in range(0,512000,step):
    if(i+dim<=512000):
        categorie_num = categorie_num+1
        x_train.append(NF[i:i+dim])

for i in range(0,512000,step):
    if(i+dim<=512000):
        x_train.append(IF[i:i+dim])
        
for i in range(0,512000,step):
    if(i+dim<=512000):
        x_train.append(OF[i:i+dim])

for i in range(0,512000,step):
    if(i+dim<=512000):
        x_train.append(BF[i:i+dim])
        
x_train=np.stack(x_train)

print(x_train.shape)
print(categorie_num)
```
